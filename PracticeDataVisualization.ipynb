```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from numpy.random import randint
```

```
#define x
x = np.linspace(0,10,20)
```

```
x
```

```
y = randint(0,50,20)
```

```
y
```

```
y = np.sort(y)
```

```
y
```

```
plt.plot(x,y, linestyle = '--', marker='*', markersize=10, lw = 1.5 )
plt.xlabel('X variable')
plt.ylabel(' Y Label')
plt.title('X vs Y Label')
plt.show()
```

```
#multiple plot on same canvas
plt.subplot(1,3,1)
plt.plot(x,y, color='r')
plt.subplot(1,3,2)
plt.plot(x,y,color='navy')
plt.subplot(1,3,3)
plt.plot(x,y,color='navy')
```

```
fig = plt.figure()
axes = fig.add_axes([0.1,0.2,1,1])
axes.plot(x,y)
axes.set_xlabel('X label axis')
axes.set_ylabel("Y label axis");
```

```
#2 sets to one canvas
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.2,0.9,0.9])
ax2 = fig.add_axes([0.5,0.22,0.4,0.3])

ax1.plot(x,y, color='r')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y AXIS')
ax1.set_title("PLOT 1")

ax2.plot(x,y, color='m')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y AXIS')
ax2.set_title("Plot 2")
```

```
#sub plots method
fig, axes = plt.subplots()
axes.plot(x,y)
```
