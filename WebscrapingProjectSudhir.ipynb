```
import bs4 as bs
import urllib.request
import pandas as pd
```

```
src = urllib.request.urlopen('https://www.bbc.com/').read()
```

```
bsoup = bs.BeautifulSoup(src, 'lxml')
```

```
print(bsoup)
```

```
print(bsoup.title)
```

```
print(bsoup.title.text)
```

```
#get the URL links from the website 
for link in bsoup.find_all('a'):
  print(link.get('href'))
```

```
#print the first paragraph using this ( p represents paragraph)
print(bsoup.p)
```

```
#final all paragraphs
print(bsoup.find_all('p'))
```

```
print(bsoup.find('p').get_text())
```

```
#want just the paragraph text
ptags = bsoup.find_all('p')
for p in ptags:
  print(p.text)
```

```
src = urllib.request.urlopen('https://www.espncricinfo.com/ci/content/player/index.html?country=6').read()
```

```
bsoup = bs.BeautifulSoup(src, 'lxml')
tbl = bsoup.find('table')
```

```
tbl_rows = tbl.find_all('tr')
for tr in tbl_rows :
  td = tr.find_all('td')
  row = [i.text for i in td]
  print(row)
```

```
saplayers = urllib.request.urlopen('https://www.espncricinfo.com/ci/content/player/index.html?country=3').read()
```

```
bsoup_sa = bs.BeautifulSoup(saplayers, 'lxml')
tbl_sa = bsoup_sa.find('table')
```
